# <u>Discover Bundles in Specific Stock Market</u>

```python
import bundleDiscovery

scenario = 1
stock_market = "London"
year = "2018"

stocks = bundleDiscovery.readDataAndGenerateCSV(scenario)
```

### Plot All Stock Market Data

```python
bundleDiscovery.plotAllData(stock_market, year, stocks)
```

### Discover Bundles

```python
bundleDiscovery.discoverBundles(scenario)
```

### Get Similar Bundles

```python
bundleDiscovery.getSimilarBundles(sort_by = "Member Similarity", t=0)
```

### Plot All Discovered Bundles

```python
bundleDiscovery.plotAllBundles(stock_market, year, stocks)
```

### Plot a Specific Bundle

```python
bundleDiscovery.plotSelectedBundle(scenario, -1, stock_market, year, stocks)
```
